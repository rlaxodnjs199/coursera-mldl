```python
def my_dot(a,b):
	x = 0
	for i in range(a.shape[0]):
		x += (a[i] * b[i])
	return x
```

```python
np.random.seed(1)
a = np.random.rand(1000000)
b = np.random.rand(1000000)

tic = time.time()
c = my_dot(a,b)
toc = time.time()

print(f"time: {(toc-tic)*1000:.4f}")

tic = time.time()
d = np.dot(a,b)
toc = time.time()
print(f"np.dot time: {(toc-tic)*1000:.4f}")
```
